```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for plotting
x1 = np.arange(100, 5000.0, 1)
y1 = np.sqrt(np.log(5000)/(x1))

plt.plot(x1, y1)
plt.show()
```

```python
x2 = np.arange(100, 5000.0, 1)
y2 = np.sqrt(np.log(x2)/(x2-1))
plt.plot(x2, y2)
plt.show()
```

```python
x3 = np.arange(100, 5000.0, 1)
y3 = np.sqrt(np.log(x3)/(1))

plt.plot(x3, y3)
plt.show()
```
